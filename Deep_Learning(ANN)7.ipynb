*1 Method*

*This is a 2D (2DIMENSION Datasets)*
x & y direction

**Flatten Layer is for Converting 2D Datasets into 1D Datasets**

```python
model=tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten(input_shape=[28,28])) #2D information into 1D array
#First Hidden Layer
model.add(tf.keras.layers.Dense(250,activation='relu'))
#Second Hidden Layer
model.add(tf.keras.layers.Dense(100,activation='relu'))
#Output  Hidden Layer
model.add(tf.keras.layers.Dense(10,activation='softmax'))
```

Method 2

```python
LAYERS = [tf.keras.layers.Flatten(input_shape=[28,28],name='InputLayer'),
          tf.keras.layers.Dense(128,activation='relu',name='HiddenLayer1'),
          tf.keras.layers.Dense(100,activation='relu',name='HiddenLayer2'),
          tf.keras.layers.Dense(10,activation='softmax',name='OutputLayer')]
```

```python
model=tf.keras.models.Sequential(LAYERS)
```

```python
model.summary()
```

*Parameters is main Disadvantages for ANN Image Classification*

```python
hidden_layer=model.layers[1]
```

```python
hidden_layer.get_weights()
```

```python
model.summary()
```

**Initial Bias will be 0**

```python
model.compile(loss ='sparse_categorical_crossentropy',
              optimizer='adam',
              metrics= ['accuracy'])
```

```python
checkpoint_callback=tf.keras.callbacks.ModelCheckpoint('my_model.h5',save_best_only=True)
```

**Taking Validation Datasets**

```python
X_valid, X_train = X_train[:5000], X_train[5000:10000]
y_valid, y_train = y_train[:5000], y_train[5000:10000]
```

```python
print("This is Validation Dataset",X_valid.shape)
print("This is Training Dataset",X_train.shape)
```

```python
model.fit(X_train,y_train,epochs=50,validation_data=(X_valid,y_valid),batch_size=8,callbacks=[checkpoint_callback])
```
