# Heart Disease Prediction Using ANN(DL)


* **Intrdouction**
  *  Heart disease prediction using Artificial Neural Networks (ANN) is a cutting-edge approach in medical diagnostics that leverages the power of machine learning to assess an individual's risk of developing cardiovascular diseases.
  * This method employs complex computational models designed to mimic the human brain's neural networks, enabling the system to learn from large datasets and make predictions based on patterns it identifies.

*  **About the data:**
   *  In the dataset, we have 13 columns in which we are given different attributes such as sex, age, cholesterol level, etc. and we are given a target column which tells us whether that person has heart disease or not.
  *  We will keep all the columns as independent variables other than the target column because it will be our dependent variable.
   *  We will build an ANN which will predict whether a person has heart disease or not given other attributes of the person.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import seaborn as sns
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("/content/heart.csv")
data
```

*Separting into Dependent & Idependent Variable*

```python
X = data.iloc[:,:13].values
y = data["target"].values
```

```python
print(X)
```

```python
print(y)
```

*Spliting into Train & Test Dataset*

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

*Building the Model*

```python
model=tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(activation='relu',input_dim=13,units=8,kernel_initializer='uniform'))
model.add(tf.keras.layers.Dense(activation='relu',units=14,kernel_initializer='uniform'))
model.add(tf.keras.layers.Dense(activation='sigmoid',units=1,kernel_initializer='uniform'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

*Fitting the Model*

```python
model.fit(X_train , y_train , batch_size = 8 ,epochs = 100)
```

```python
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)
```

*Confusion Matrix*

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

*Checking Accuracy*

```python
accuracy = (cm[0][0]+cm[1][1])/(cm[0][1] + cm[1][0] +cm[0][0] +cm[1][1])  #Accuracy Formula
print(accuracy*100)
```

We will get accuracy approximately around 87&
